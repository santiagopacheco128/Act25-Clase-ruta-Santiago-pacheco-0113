* 3 atributos (origen, destino, distancia)
* 2 funciones leerdatos(), mostrardatos()
* 2 objetos (ruta1, ruta2)

![texto del vínculo](https://raw.githubusercontent.com/santiagopacheco128/Act25-Clase-ruta-Santiago-pacheco-0113/refs/heads/main/Act25%20clase%20ruta%20Santiago%200113.png)

```python
class Ruta:
    def __init__(self, origen, destino, distancia):
        # Atributos iniciales
        self.origen = origen
        self.destino = destino
        self.distancia = distancia

    def leerdatos(self):
        print(f"\n[SISTEMA]: 📍 Leyendo datos de la ruta...")
        print(f"Origen: {self.origen} | Destino: {self.destino} | Distancia: {self.distancia} km")

    def mostrardatos(self):
        print(f"\n[SISTEMA]: 🗺️ Mostrando información completa:")
        print(f"Ruta: {self.origen} a {self.destino} ({self.distancia} km)")

# --- Creación de 2 objetos Ruta ---
ruta1 = Ruta("Ciudad de México", "Cancún", 1600)
ruta2 = Ruta("Madrid", "Barcelona", 620)

# --- Menú Interactivo ---
while True:
    print("\n" + "—" * 30)
    print("      MENÚ PRINCIPAL")
    print("—" * 30)
    print(f"1. Usar ruta {ruta1.origen}")
    print(f"2. Usar ruta {ruta2.origen}")
    print("3. Salir del programa")

    ruta_op = input("\nElige una ruta (1-3): ")

    if ruta_op == "3":
        print("\nSaliendo del simulador...")
        break

    if ruta_op == "1":
        obj = ruta1
    elif ruta_op == "2":
        obj = ruta2
    else:
        print("Opción no válida.")
        continue

    print(f"\n>>> TRABAJANDO CON: {obj.origen} <<<")
    print(f"Atributos: Destino {obj.destino} | Distancia: {obj.distancia}")
    print("a. Leer Datos")
    print("b. Mostrar Datos")

    accion = input("Elige una acción (a o b): ").lower()

    if accion == "a":
        obj.leerdatos()
    elif accion == "b":
        obj.mostrardatos()
    else:
        print("Acción no reconocida.")

print("  Programa realizado por Santiago Pacheco NL = 45  ")
```
